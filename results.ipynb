### Result Visualization

##### Sequence Labelling Baseline

```python
import pandas as pd
import os

sl_base_val_path = os.path.join("run", "sequence-labelling-ss1-e10-lr4e-4-use_weighted_loss-freeze.base-2a4z0hia", "validation_log.9.csv")
df = pd.read_csv(sl_base_val_path)
df.head(5)
```

```python
from utils.seqlab import Index_Dictionary
for i in range(8):
    label = Index_Dictionary[i]
    df[f"precision-{label}"] = 0
    df[f"recall-{label}"] = 0
df.head(5)
```

```python
from tqdm import tqdm
from utils.metrics import Metrics, cleanup_prediction
from utils.seqlab import Index_Dictionary
import numpy as np

precisions = {}
recalls = {}
for j in range(8):
    label = Index_Dictionary.get(j)
    precisions[f"{label}"] = []
    recalls[f"{label}"] = []

for i, r in df.iterrows():
    pred = [int(a) for a in r["prediction"].split(' ')]
    target = [int(a) for a in r["target"].split(' ')]
    cp, ct = cleanup_prediction(pred, target)
    metric = Metrics(cp, ct)
    metric.calculate()

    for j in range(8):
        label = Index_Dictionary[j]
        pres = metric.precision(j)
        rec = metric.recall(j)
        print(f"{pres}, {rec}                     ", end='\r')
        precisions[f"{label}"].append(pres)
        recalls[f"{label}"].append(rec)

for j in range(8):
    label = Index_Dictionary[j]
    df[f"precision-{label}"] = precisions[f"{label}"]
    df[f"recall-{label}"] = recalls[f"{label}"]

df.head(10)
```
